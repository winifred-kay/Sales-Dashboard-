## Data Cleaning

```python
#Changing the column names of the first half of the year 
Jan.rename(columns = {'Order ID': 'Order_ID','Quantity Ordered':'Quantity_Ordered','Price Each':'Price_Each','Order Date':'Order_Date','Purchase Address':'Purchase_Address'}, inplace=True)
Feb.rename(columns = {'Order ID': 'Order_ID','Quantity Ordered':'Quantity_Ordered','Price Each':'Price_Each','Order Date':'Order_Date','Purchase Address':'Purchase_Address'}, inplace=True)
Mar.rename(columns = {'Order ID': 'Order_ID','Quantity Ordered':'Quantity_Ordered','Price Each':'Price_Each','Order Date':'Order_Date','Purchase Address':'Purchase_Address'}, inplace=True)
Apr.rename(columns = {'Order ID': 'Order_ID','Quantity Ordered':'Quantity_Ordered','Price Each':'Price_Each','Order Date':'Order_Date','Purchase Address':'Purchase_Address'}, inplace=True)
May.rename(columns = {'Order ID': 'Order_ID','Quantity Ordered':'Quantity_Ordered','Price Each':'Price_Each','Order Date':'Order_Date','Purchase Address':'Purchase_Address'}, inplace=True)
Jun.rename(columns = {'Order ID': 'Order_ID','Quantity Ordered':'Quantity_Ordered','Price Each':'Price_Each','Order Date':'Order_Date','Purchase Address':'Purchase_Address'}, inplace=True)
```

```python
# Changing the Data Types for Order_ID, Quantity_Ordered and Price_Each
Jan['Order_ID'] = pd.to_numeric(Jan['Order_ID'], errors='coerce')
Jan['Quantity_Ordered'] = pd.to_numeric(Jan['Quantity_Ordered'], errors='coerce')
Jan['Price_Each'] = pd.to_numeric(Jan['Price_Each'], errors='coerce')

Feb['Order_ID'] = pd.to_numeric(Feb['Order_ID'], errors='coerce')
Feb['Quantity_Ordered'] = pd.to_numeric(Feb['Quantity_Ordered'], errors='coerce')
Feb['Price_Each'] = pd.to_numeric(Feb['Price_Each'], errors='coerce')

Mar['Order_ID'] = pd.to_numeric(Mar['Order_ID'], errors='coerce')
Mar['Quantity_Ordered'] = pd.to_numeric(Mar['Quantity_Ordered'], errors='coerce')
Mar['Price_Each'] = pd.to_numeric(Mar['Price_Each'], errors='coerce')

Apr['Order_ID'] = pd.to_numeric(Apr['Order_ID'], errors='coerce')
Apr['Quantity_Ordered'] = pd.to_numeric(Apr['Quantity_Ordered'], errors='coerce')
Apr['Price_Each'] = pd.to_numeric(Apr['Price_Each'], errors='coerce')

May['Order_ID'] = pd.to_numeric(May['Order_ID'], errors='coerce')
May['Quantity_Ordered'] = pd.to_numeric(May['Quantity_Ordered'], errors='coerce')
May['Price_Each'] = pd.to_numeric(May['Price_Each'], errors='coerce')

Jun['Order_ID'] = pd.to_numeric(Jun['Order_ID'], errors='coerce')
Jun['Quantity_Ordered'] = pd.to_numeric(Jun['Quantity_Ordered'], errors='coerce')
Jun['Price_Each'] = pd.to_numeric(Jun['Price_Each'], errors='coerce')
```

```python
Jan['Sales_Month'] = 'January'
Feb['Sales_Month'] = 'February'
Mar['Sales_Month'] = 'March'
Apr['Sales_Month'] = 'April'
May['Sales_Month'] = 'May'
Jun['Sales_Month'] = 'June'
Jul['Sales_Month'] = 'July'
Aug['Sales_Month'] = 'August'
Sep['Sales_Month'] = 'September'
Oct['Sales_Month'] = 'October'
Nov['Sales_Month'] = 'November'
Dec['Sales_Month'] = 'December'
```

```python
#change the date column to datetime 
Feb = Feb[Feb['Order_Date'] != 'Order Date']
Feb['Order_Date'] = pd.to_datetime(Feb['Order_Date'])

Sep = Sep[Sep['Order_Date'] != 'Order Date']
Sep['Order_Date'] = pd.to_datetime(Sep['Order_Date'])
```

```python
full_year = pd.concat([Jan,Feb,Mar,Apr,May,Jun,Jul,Aug,Sep,Oct,Nov,Dec])
```

```python
full_year.info()
full_year.head()
```

```python
# Assuming you have a DataFrame called 'df'
full_year = full_year.drop(columns=full_year.columns[-1])
```

```python
# Create 'Product_Level' column based on 'Price_Each'
full_year['Product_Level'] = full_year['Price_Each'].apply(lambda price: 'High-Level Product' if price > 99.99 else 'Basic-Level Product')
```

```python
full_year = full_year.dropna()
```
